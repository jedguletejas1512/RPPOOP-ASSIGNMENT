```python
class BankAccount:
    def __init__(self, name, account_number, account_type, balance=0):
        self.name = name
        self.account_number = account_number
        self.account_type = account_type
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Amount {amount} deposited into account {self.account_number}.")
    
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Amount {amount} withdrawn from account {self.account_number}.")
        else:
            print("\n \033[1m  !!** Insufficient balance. **!! \033[0m")
    
    def display_details(self):
        print("\033[1m   ******************Account Details***********************  \033[0m")
        print("\033[1m   Name:  \033[0m", self.name)
        print("\033[1m   Account Number:  \033[0m", self.account_number)
        print("\033[1m   Account Type:  \033[0m", self.account_type)
        print("\033[1m   Balance:  \033[0m", self.balance)
        print("\033[1m   *****************************************  \033[0m")
        



accounts = {}

def create_account():
    name = input("\033[1m   Enter name of account holder:  \033[0m")
    account_number = input("\033[1m   Enter account number:  \033[0m")
    account_type = input("\033[1m   Enter account type (savings or current):  \033[0m")
    balance = float(input("\033[1m   Enter initial balance:  \033[0m"))
#     account = BankAccount(name, acc_no, acc_type, balance)
      
    
    if account_number in accounts:
        print("**********Account number already exists. Please choose a different account number.**********")
        return

    account = BankAccount(name, account_number, account_type, balance)
    accounts[account_number] = account
    print("\033[1m------Account created successfully.------\033[0m")  


def deposit_amount():
    account_number = input("\033[1m Enter account number: \033[0m ")
    if account_number in accounts:
        account = accounts[account_number]
        amount = float(input("\033[1m Enter amount to deposit: \033[0m"))
        account.deposit(amount)
    else:
        print("\033[1m ------- Account does not exist-------\033[0m ")


def withdraw_amount():
    account_number = input("\033[1m Enter account number: \033[0m ")
    if account_number in accounts:
        account = accounts[account_number]
        amount = float(input("\033[1m Enter amount to withdraw: \033[0m"))
        account.withdraw(amount)
    else:
        print("\033[1m ------- Account does not exist-------\033[0m ")


def display_account_details():
    account_number = input("\033[1m Enter account number: \033[0m ")
    if account_number in accounts:
        account = accounts[account_number]
        account.display_details()
    else:
        print("\033[1m ------- Account does not exist-------\033[0m ")


def menu():
    while True:
        
        print("\n\033[1m   Welcome to the Bank Account Management   \033[0m\n")
        print("\033[1m  1. Create new account  \033[0m")
        print("\033[1m  2. Deposit amount  \033[0m")
        print("\033[1m  3. Withdraw amount  \033[0m")
        print("\033[1m  4. Display account information  \033[0m")
        print("\033[1m  5. Exit\n  \033[0m")
        choice = int(input("Enter your choice (1-5):  "))


    
        if choice == 1:
            create_account()
        elif choice == 2:
            deposit_amount()
        elif choice == 3:
            withdraw_amount()
        elif choice == 4:
            display_account_details()
        elif choice == 5:
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    menu()
```
